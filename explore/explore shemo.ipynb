## create dataframe for dataset

```python
!mkdir shemo
!mv ./male/* ./shemo/
!mv ./female/* ./shemo/
```

```python
!rm ./male -r
!rm ./female -r
```

```python
import glob

files = glob.glob('shemo/*')
```

```python
print(files)
```

```python
import pandas as pd
df = pd.DataFrame({"path" : files})
```

```python
df
```

```python
df['file'] = df['path'].apply(lambda x: x.split('/')[-1])
```

```python
def find_gender(file):
  if file[0] == 'M':
    return 'male'
  return 'female'

df['gender'] = df['file'].apply(find_gender)
```

```python
def find_speaker_id(file):
  return file[:3]

df['speaker_id'] = df['file'].apply(find_speaker_id)
```

```python
df
```

```python
from scipy.io import wavfile
def find_dur(path):
  sample_rate, data = wavfile.read(path)
  len_data = len(data)
  secs = len_data / sample_rate
  return secs
```

```python
df['duration'] = df['path'].apply(find_dur)
```

```python
df
```

```python
df[df['duration']< 1]
```

```python
df[df['duration']> 10]
```

```python
df = df[(df['duration'] > 1) & (df['duration'] < 10)]
```

```python
df
```

## split to train, validation, and test

```python
import numpy as np

def split_train_test(speakers):
  np.random.seed(42)
  shuffle_indices = np.random.permutation(len(speakers))
  speakers = speakers[shuffle_indices]
  sz = len(speakers)
  val_index = int(0.8*sz)
  test_index = int(0.9*sz)
  train, validation, test = speakers[:val_index], speakers[val_index:test_index], speakers[test_index:]

  return train, validation, test
```

```python
male_train, male_validation, male_test = split_train_test(male_speakers)
```

```python
female_train, female_validation, female_test = split_train_test(female_speakers)
```

```python
train = df[df['speaker_id'].isin(male_train) | df['speaker_id'].isin(female_train)]
```

```python
train
```

```python
validation = df[df['speaker_id'].isin(male_validation) | df['speaker_id'].isin(female_validation)]
```

```python
validation
```

```python
test = df[df['speaker_id'].isin(male_test) | df['speaker_id'].isin(female_test)]
```

```python
test
```
